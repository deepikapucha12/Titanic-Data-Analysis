```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/archive (2).zip")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df["Age"] = df["Age"].fillna(df["Age"].mean(), inplace=True)
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0], inplace=True)
```

```python
df["Survived"].value_counts()
```

```python
gender_survival = df.groupby("Sex")["Survived"].sum()
gender_survival
```

```python
class_survival = df.groupby("Pclass")["Survived"].sum()
class_survival
```

```python
df["Survived"].value_counts().plot(kind="bar")
plt.title("Survival Count")
plt.xlabel("Survived (0= No, 1 = Yes)")
plt.ylabel("Count")
plt.show()
```

```python
gender_survival.plot(kind="bar")
plt.title("Survival by Gender")
plt.xlabel("Gender")
plt.ylabel("Survived Count")
plt.show()
```

```python
class_survival.plot(kind="bar")
plt.title("Survival by Class")
plt.xlabel("Class")
plt.ylabel("Survived Count")
plt.show()
```

```python
import seaborn as sns
sns.countplot(x="Sex", hue="Survived", data=df)
plt.show()
```

```python
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True)
plt.show()
```

Insights:
* Females had higher survival rate
* First class passengers survived more
* Survival depends on multiple factors
